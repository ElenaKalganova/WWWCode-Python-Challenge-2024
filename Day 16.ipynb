Write a function that counts the frequency of each word in a sentence.

```python
def count_words(text):
    """Counts the frequency of each word in a string.

    Parameters
    ----------
    text: str
        The text in which to count words.

    Returns
    -------
    dict
        A dictionary of unique words in the text and their frequency.
    """
    unique_words={}
    all_words=text.lower().split()
    for word in all_words:
        if word in unique_words:
            unique_words[word]+=1
        else:
            unique_words[word]=1
    return(sorted(unique_words.items(), key=lambda x:x[1], reverse=True))
```
